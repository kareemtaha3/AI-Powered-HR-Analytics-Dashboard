# Employee Skills Clustering - Data Exploration

This notebook performs comprehensive exploratory data analysis (EDA) on the CareerMap dataset which maps tech roles with personality traits and skills.

## 1. Import Libraries

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set visualization style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
%matplotlib inline

print("Libraries imported successfully!")
```

## 2. Load the Dataset

```python
# Load the dataset
df = pd.read_csv('../Data/CareerMap- Mapping Tech Roles With Personality & Skills.csv')

print(f"Dataset loaded successfully!")
print(f"Dataset shape: {df.shape}")
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
```

## 3. Initial Data Inspection

```python
# Display first few rows
print("First 5 rows of the dataset:")
print("="*100)
df.head()
```

```python
# Display dataset information
print("Dataset Information:")
print("="*100)
df.info()
```

```python
# Display column names
print("Column Names:")
print("="*100)
for i, col in enumerate(df.columns, 1):
    print(f"{i}. {col}")
```

```python
# Statistical summary
print("Statistical Summary:")
print("="*100)
df.describe()
```

## 4. Data Quality Check

```python
# Check for missing values
print("Missing Values Analysis:")
print("="*100)
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
missing_df = pd.DataFrame({
    'Column': missing_values.index,
    'Missing Count': missing_values.values,
    'Percentage': missing_percentage.values
})
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing Count', ascending=False)

if len(missing_df) == 0:
    print("✓ No missing values found in the dataset!")
else:
    print(missing_df)
```

```python
# Check for duplicates
print("Duplicate Rows Analysis:")
print("="*100)
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
print(f"Percentage of duplicates: {(duplicates/len(df)*100):.2f}%")
```

## 5. Target Variable Analysis (Role Distribution)

```python
# Analyze Role distribution
print("Role Distribution:")
print("="*100)
role_counts = df['Role'].value_counts()
print(role_counts)
print(f"\nTotal unique roles: {df['Role'].nunique()}")
```

```python
# Visualize Role distribution
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Bar plot
role_counts.plot(kind='bar', ax=axes[0], color='skyblue', edgecolor='black')
axes[0].set_title('Distribution of Tech Roles', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Role', fontsize=12)
axes[0].set_ylabel('Count', fontsize=12)
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(axis='y', alpha=0.3)

# Pie chart
axes[1].pie(role_counts, labels=role_counts.index, autopct='%1.1f%%', startangle=90)
axes[1].set_title('Role Distribution (Percentage)', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

## 6. Skills Analysis

```python
# Identify skill columns (excluding personality traits and Role)
skill_columns = df.columns[:17].tolist()  # First 17 columns are skills
print("Skill Columns:")
print("="*100)
for i, skill in enumerate(skill_columns, 1):
    print(f"{i}. {skill}")
```

```python
# Distribution of skill levels
fig, axes = plt.subplots(4, 5, figsize=(20, 16))
axes = axes.flatten()

for idx, skill in enumerate(skill_columns):
    if idx < len(axes):
        df[skill].value_counts().sort_index().plot(kind='bar', ax=axes[idx], color='coral', edgecolor='black')
        axes[idx].set_title(f'{skill}', fontsize=10, fontweight='bold')
        axes[idx].set_xlabel('Skill Level', fontsize=9)
        axes[idx].set_ylabel('Count', fontsize=9)
        axes[idx].tick_params(axis='x', rotation=0)
        axes[idx].grid(axis='y', alpha=0.3)

# Hide extra subplots
for idx in range(len(skill_columns), len(axes)):
    axes[idx].axis('off')

plt.suptitle('Distribution of Skill Levels', fontsize=16, fontweight='bold', y=1.00)
plt.tight_layout()
plt.show()
```

```python
# Average skill levels
avg_skills = df[skill_columns].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
avg_skills.plot(kind='barh', color='teal', edgecolor='black')
plt.title('Average Skill Levels Across All Employees', fontsize=14, fontweight='bold')
plt.xlabel('Average Skill Level', fontsize=12)
plt.ylabel('Skills', fontsize=12)
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

print("\nAverage Skill Levels:")
print("="*100)
print(avg_skills)
```

## 7. Personality Traits Analysis

```python
# Identify personality trait columns
personality_columns = df.columns[17:-1].tolist()  # Columns between skills and Role
print("Personality Trait Columns:")
print("="*100)
for i, trait in enumerate(personality_columns, 1):
    print(f"{i}. {trait}")
```

```python
# Distribution of personality traits
fig, axes = plt.subplots(3, 4, figsize=(18, 12))
axes = axes.flatten()

for idx, trait in enumerate(personality_columns):
    if idx < len(axes):
        axes[idx].hist(df[trait], bins=30, color='lightgreen', edgecolor='black', alpha=0.7)
        axes[idx].set_title(f'{trait}', fontsize=10, fontweight='bold')
        axes[idx].set_xlabel('Value', fontsize=9)
        axes[idx].set_ylabel('Frequency', fontsize=9)
        axes[idx].grid(axis='y', alpha=0.3)

# Hide extra subplots
for idx in range(len(personality_columns), len(axes)):
    axes[idx].axis('off')

plt.suptitle('Distribution of Personality Traits', fontsize=16, fontweight='bold', y=1.00)
plt.tight_layout()
plt.show()
```

```python
# Statistical summary of personality traits
print("Statistical Summary of Personality Traits:")
print("="*100)
df[personality_columns].describe()
```

## 8. Correlation Analysis

```python
# Correlation matrix for skills
plt.figure(figsize=(14, 10))
correlation_matrix_skills = df[skill_columns].corr()
sns.heatmap(correlation_matrix_skills, annot=False, cmap='coolwarm', center=0, 
            linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix - Skills', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix for personality traits
plt.figure(figsize=(12, 8))
correlation_matrix_personality = df[personality_columns].corr()
sns.heatmap(correlation_matrix_personality, annot=True, cmap='viridis', center=0, 
            linewidths=0.5, cbar_kws={"shrink": 0.8}, fmt='.2f')
plt.title('Correlation Matrix - Personality Traits', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 9. Box Plots for Outlier Detection

```python
# Box plots for personality traits to detect outliers
fig, axes = plt.subplots(3, 4, figsize=(18, 12))
axes = axes.flatten()

for idx, trait in enumerate(personality_columns):
    if idx < len(axes):
        axes[idx].boxplot(df[trait], vert=True)
        axes[idx].set_title(f'{trait}', fontsize=10, fontweight='bold')
        axes[idx].set_ylabel('Value', fontsize=9)
        axes[idx].grid(axis='y', alpha=0.3)

# Hide extra subplots
for idx in range(len(personality_columns), len(axes)):
    axes[idx].axis('off')

plt.suptitle('Box Plots - Personality Traits (Outlier Detection)', fontsize=16, fontweight='bold', y=1.00)
plt.tight_layout()
plt.show()
```
